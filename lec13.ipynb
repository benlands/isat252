```python
import pandas
```

```python
df = pandas.read_excel("s3://buckethead252/house_price.xls") 
df.plot(x="built_in",y ="price")
```

## line cahrt

# avg prices per year

```python
avg_price_per_year = df.groupby('built_in').mean()["price"]
avg_price_per_year.plot()
```

## histogram

```python
df["price"].hist()
```

most houses are around $200K

avg prices per house type

```python
avg_price_house_type = df.groupby("house_type").mean()['price']
avg_price_house_type.plot.bar()
```

# Pie chart

```python
count_per_house_type = df["house_type"].value_counts()
count_per_house_type.plot.pie()
```

## scatter plot

the relationship between price and area

```python
df.plot.scatter(x="area",y = "price")
```
